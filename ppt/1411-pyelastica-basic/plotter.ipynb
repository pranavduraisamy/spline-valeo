```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import warnings
warnings.simplefilter('ignore')
lsls=[]
lstb=[]

def length(x):
    df=pd.DataFrame(x,columns=['x','y','z'])
    disc_dst=np.sqrt((df['x'].diff()**2) + (df['y'].diff()**2) + (df['z'].diff()**2))
    lngth=disc_dst.sum()
    return lngth

def plotter(psit=1,case=1):
    df=pd.read_csv('../../data/2810/p'+str(psit)+'-case'+str(case)+'.csv')
    prt=df.iloc[10:,2:].reset_index(drop=True).values
    p0=df.iloc[5,2:].values.astype(float)   
    p1=df.iloc[6,2:].values.astype(float)   
    u0=df.iloc[8,2:].values.astype(float)*df.iloc[2,1]   
    u1=df.iloc[9,2:].values.astype(float)*df.iloc[3,1]   
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(x=prt[:,0],y=prt[:,1],z=prt[:,2],mode='lines',name='Prototype',line=dict(color='darkslategray',width=6)))
    fig.add_trace(go.Scatter3d(x=[p0[0]],y=[p0[1]],z=[p0[2]],mode='markers',name='P0',marker=dict(size=5,color='red')))
    fig.add_trace(go.Scatter3d(x=[p1[0]],y=[p1[1]],z=[p1[2]],mode='markers',name='P1',marker=dict(size=5,color='red')))
    fig.add_trace(go.Cone(x=[p0[0]],y=[p0[1]],z=[p0[2]],u=[u0[0]],v=[u0[1]],w=[u0[2]],colorscale='RdPu',sizemode='absolute',sizeref=6,showscale=False,name='Unit vector at P0',showlegend=True))
    fig.add_trace(go.Cone(x=[p1[0]],y=[p1[1]],z=[p1[2]],u=[u1[0]],v=[u1[1]],w=[u1[2]],colorscale='RdPu',sizemode='absolute',sizeref=6,showscale=False,name='Unit vector at P1',showlegend=True))

    a=df.iloc[59,2:]

    #100p base
    spl3=pd.read_csv('../../pyelastica/data/p'+str(psit)+'-case'+str(case)+'.csv')
    errb=length(pd.DataFrame([a,spl3.iloc[50]]).values).round(3)
    fig.add_trace(go.Scatter3d(x=[a[0],spl3.iloc[50,0]],y=[a[1],spl3.iloc[50,1]],z=[a[2],spl3.iloc[50,2]],mode='lines',name='Error',line=dict(color='red',width=6,dash='longdash')))
    spl3=spl3.values
    fig.add_trace(go.Scatter3d(x=spl3[:,0],y=spl3[:,1],z=spl3[:,2],mode='lines',name='Simulated',line=dict(color='darkorchid',width=6)))
    fig.update_layout(title='Simulated - '+'P'+str(psit)+'C'+str(case),scene=dict(xaxis_title='X',yaxis_title='Y',zaxis_title='Z',camera=dict(eye=dict(x=1,y=-1,z=1))),font=dict(size=15,family='Times New Roman'),autosize=False,height=550,width=900,legend=dict(x=0.02,y=0.98),margin=dict(l=20,r=20,t=40,b=20))
    fig.update_scenes(camera_projection_type='orthographic')
    fig.show()
    ln_spl=length(spl3)-length(prt)

    return errb,ln_spl
```

```python
pos=[]
cas=[]
prt_len=[]
herm_mder=[]
herm_lener=[0.0]*12
for i in range(1,5):
    for j in range(1,4):
        pos.append(int(i))
        cas.append(int(j))
        prt_len.append(length(pd.read_csv('../../data/2810/p'+str(i)+'-case'+str(j)+'.csv').iloc[10:,2:].values))
        temp1=pd.read_csv('../../data/2810/p'+str(i)+'-case'+str(j)+'.csv').iloc[59,2:]
        temp2=pd.read_csv('../../data/2810/opt/p'+str(i)+'-case'+str(j)+'-opt.csv').iloc[50]
        herm_mder.append(length(pd.DataFrame([temp1,temp2]).values))
df=pd.DataFrame([pos,cas,prt_len,herm_mder,lstb,herm_lener,lsls]).T
df.columns=['Position','Case','Length - Prototype','MD Error - Hermite','MD Error - Simulation','Length Error - Hermite','Length Error - Simulation']
df['Position']=df['Position'].astype(int)
df['Case']=df['Case'].astype(int)
df.round(3)
```
