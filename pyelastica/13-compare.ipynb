```python
df=[]
si=[]
pos=[]
cas=[]
ln=[]
for i in range(1,5):
    for j in range(1,4):
        df.append(pd.read_csv('../data/2810/p'+str(i)+'-case'+str(j)+'.csv'))
        si.append(pd.read_csv('data/p'+str(i)+'-case'+str(j)+'.csv'))
        ln.append(utils.length(pd.DataFrame([df[len(pos)].iloc[59,2:],si[len(pos)].iloc[50]]).values))
        pos.append(i)
        cas.append(j)
cor=pd.DataFrame([pos,cas,ln]).T
cor.columns=['Position','Case','Error']
cor
```
