```python
import pandas as pd
import utils
```

```python
df=[]
si=[]
he=[]
pos=[]
cas=[]
er_sim=[]
er_her=[]
for i in range(1,5):
    for j in range(1,4):
        df.append(pd.read_csv('../data/2810/p'+str(i)+'-case'+str(j)+'.csv'))
        si.append(pd.read_csv('data/p'+str(i)+'-case'+str(j)+'.csv'))
        he.append(pd.read_csv('../data/2810/opt/p'+str(i)+'-case'+str(j)+'-opt.csv'))
        er_sim.append(utils.length(pd.DataFrame([df[len(pos)].iloc[59,2:],si[len(pos)].iloc[50]]).values))
        er_her.append(utils.length(pd.DataFrame([df[len(pos)].iloc[59,2:],he[len(pos)].iloc[50]]).values))
        pos.append(int(i))
        cas.append(int(j))
cor=pd.DataFrame([pos,cas,er_sim,er_her]).T
cor.columns=['Position','Case','Error - Simulation','Error - Hermite']
cor
```
