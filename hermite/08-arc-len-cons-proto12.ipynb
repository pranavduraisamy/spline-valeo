# ***Description***

### Arc length constrained hermite spline with all the 12 cases given on 28 Oct 25

```python
import utils
from scipy import optimize
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
import warnings
warnings.simplefilter('ignore')
mag=[]
prt_len=[]
ach_len=[]
err_len=[]
```

# Position 1

## Case 01

```python
df=pd.read_csv('../data/2810/p1-case1.csv')
prt=df.iloc[10:,2:].reset_index(drop=True)

p0=df.iloc[5,2:].values.astype(float)    # Point0
p1=df.iloc[6,2:].values.astype(float)    # Point1
u0=df.iloc[8,2:].values.astype(float)    # Unit Vector @ Point0
u1=df.iloc[9,2:].values.astype(float)    # Unit Vector @ Point1
m0=df.iloc[2,1]                          # Magnitude for Vector1
m1=df.iloc[3,1]                          # Magnitude for Vector2
u0_init=u0*m0
u1_init=u1*m1
spl=utils.hermite(p0, p1, u0_init, u1_init,num_points=51)

tgt=utils.length(prt.values)             # Target spline length

obj_fn=lambda x: utils.length(utils.hermite(p0,p1,x*u0,x*u1)) - tgt
sol=optimize.root_scalar(obj_fn, bracket=[0, tgt*3], method='brentq')
solx=sol.root
u0_final=solx*u0
u1_final=solx*u1
spl=utils.hermite(p0,p1,u0_init,u1_init,num_points=100)
spl_tgt=utils.hermite(p0,p1,u0_final,u1_final,num_points=100)

print(f'Length of the Prototype Spline: {utils.length(prt.values):.5f}')
print(f'Given target spline length: {tgt:.5f}')
print(f'Length of the Hermite Spline: {utils.length(spl.astype(float)):.5f}')
print(f'Achieved Length of the Hermite Spline: {utils.length(spl_tgt.astype(float)):.5f}')
print(f'\nComputed Magnitude (X): {solx:.4f}')

a=prt.iloc[50]
b=pd.DataFrame(spl_tgt,columns=prt.columns).iloc[50]
mag.append(np.round(solx,3))
prt_len.append(utils.length(prt.values).round(3))
ach_len.append(utils.length(spl_tgt.astype(float)).round(3))
err_len.append(utils.length(pd.DataFrame([a,b]).values).round(3))
fig=utils.plotter_v2(spl,p0,p1,u0_init,u1_init,spl1=spl_tgt,spl2=prt.values,x_label='Unoptimized Hermite Spline',
                     spl1_label='Optimized hermite spline',spl2_label='Prototype',cne=6,intr='d',height=550,width=900,title='Position 1 Case 1')
fig.add_trace(go.Scatter3d(x=[a[0],b[0]],y=[a[1],b[1]],z=[a[2],b[2]],mode='lines',name='Midpoint Deviation'
                           ,line=dict(color='red',width=6,dash='longdash'),text='Length '+str(utils.length(pd.DataFrame([a,b]).values).round(2)))) 
fig.update_layout(legend=dict(x=0.75,y=1,xanchor="left",yanchor="top"))
fig.update_layout(autosize=True)
pio.write_html(fig,'../ppt/1411-arc-len-herm/p1-case1.html')
fig.show()
```
