# Position 1

## P1 Case 01

```python
from EasyFEA import Display,Models,plt,Mesher,ElemType,Simulations
from EasyFEA.Geoms import Circle,Line
import pandas as pd
import utils 
Display.Clear()

# Import
df=pd.read_csv('../data/2810/opt/p1-case1-opt.csv')
df=df.values

# CS
circle_ot=Circle((0,0),6,0.2)
circle_in=Circle((0,0),3.5,0.25,isHollow=True)
section=circle_ot.Mesh_2D([circle_in],ElemType.QUAD8)
Display.Plot_Mesh(section)

# Spline
beam=[]
for i in range(len(df)-1):
    p1=df[i]
    p2=df[i+1]
    line=Line(p1,p2)
    el=Models.BeamElasIsot(3,line,section,10,0.49)
    beam.append(el)

mesh=Mesher().Mesh_Beams(beams=beam,elemType=ElemType.SEG2)

spline=Models.BeamStructure(beam)


load=-1*((utils.length(df)*section.area*1.23*9.81*1e-6))

# BC
simu=Simulations.BeamSimu(mesh,spline)
simu.add_dirichlet(mesh.Nodes_Point(df[0]),[0] * 6,['x','y','z','rx','ry','rz'])
simu.add_dirichlet(mesh.Nodes_Point(df[-1]),[0] * 6,['x','y','z','rx','ry','rz'])
simu.add_lineLoad(mesh.Nodes_Points(df[1:-1]),[load/len(df[1:-1])],['z'])

Display.Plot_BoundaryConditions(simu)

for i  in range(1,len(df)-1):
    simu.add_connection_fixed(mesh.Nodes_Point(df[i]))

# Solve the beam problem and get displacement results
sol=simu.Solve()
simu.Save_Iter()

# Results
Display.Plot_Mesh(simu,1)

pd.DataFrame(simu.Results_displacement_matrix(),columns=['X','Y','Z']).to_csv('../data/2810/grav/p1-case1-grav.csv',index=False)

print(simu)

plt.show()
```
