### Visulaization of optimized Hermite spline for P1 case studies. Equal Length constraint used.

## Case 01

```python
import pandas as pd
import utils

df=pd.read_csv('../data/2810/p1-case1.csv')
prt=df.iloc[10:,2:].reset_index(drop=True)

p0=df.iloc[5,2:].values.astype(float)    # Point0
p1=df.iloc[6,2:].values.astype(float)    # Point1
u0=df.iloc[8,2:].values.astype(float)    # Unit Vector @ Point0
u1=df.iloc[9,2:].values.astype(float)    # Unit Vector @ Point1
m0=df.iloc[2,1]                          # Magnitude for Vector1
m1=df.iloc[3,1]                          # Magnitude for Vector2
u0_init=u0*m0
u1_init=u1*m1
spl=utils.hermite(p0,p1,u0_init,u1_init,num_points=51)
spl_tgt=pd.read_csv('../data/2810/p1-case1-opt.csv').values
siml=pd.read_csv('../data/2810/p1-case1-grav.csv')[::2].reset_index(drop=True)
siml.loc[len(siml)]=[0,0,0]
siml=spl_tgt+siml

utils.plotter(spl,p0,p1,u0_init,u1_init,tgt=spl_tgt,prt=prt.values,siml=siml.values,cne=6,intr='s',tgt_label='Optimized hermite spline')
```
