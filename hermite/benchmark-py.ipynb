## A Hermite cubic curve has end points P0(1,3) and P1(7,2). The tangent vector at P0 is defined by line joining P0 and point P2(10,8) whereas tangent vector at P1 is defined by line joining P1 and P3(6,0). Compute the intermediate points on the curve at u=.25

```python
import utils
import numpy as np

p0=np.array([1,3,0])   # Point0
p1=np.array([7,2,0])   # Point1
u0=np.array([9,5,0])    # Unit Vector @ Point0
u1=np.array([-1,-2,0])    # Unit Vector @ Point1
m=1                     # Magnitude
u0=u0*m
u1=u1*m
spl=utils.hermite(p0, p1, u0, u1,num_points=100)

print(f'Length of the curve: {utils.length(spl):.5f}')

utils.plotter(spl,p0,p1,u0,u1)
```
