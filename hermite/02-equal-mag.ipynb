# ***Description***

## *Given,*

### Boundary points (P0, P1), Unit vectors (U0, U1) and Spline length

## *Does,*

- ### Assume magnitude of both the vectors are equal
- ### Computes a magnitude to match the given spline length
- ### Constructs and plots the hermite spline which matches the given conditions

```python
import numpy as np
from scipy import optimize
import utils

p0=np.array([7,10,0])   # Point0
p1=np.array([20,3,5])   # Point1
u0=np.array([1,1,2])    # Unit Vector @ Point0
u1=np.array([2,0,1])    # Unit Vector @ Point1

tgt=float(input('Target length for the curve: '))   # Input target spline length

obj_fn=lambda x: utils.length(utils.hermite(p0,p1,x*u0,x*u1)) - tgt
sol=optimize.root_scalar(obj_fn, bracket=[0, tgt*3], method='brentq')
solx=sol.root

u0_final=solx*u0
u1_final=solx*u1
spl_initl=utils.hermite(p0,p1,u0,u1)
spl_final=utils.hermite(p0,p1,u0_final,u1_final)

print(f'\nComputed Magnitude (X): {solx:.4f}')
print('Initial length          : ',utils.length(spl_initl))
print('Achieved target length  : ',utils.length(spl_final))

utils.plotter(spl_final,p0,p1,u0_final,u1_final)
```
