## Case 01

```python
import pandas as pd
import utils

df=pd.read_csv('../data/2810/p1-case1.csv')
prt=df.iloc[10:,2:].reset_index(drop=True)

df1=pd.read_csv('../data/2810/p2-case1.csv')
prt1=df1.iloc[10:,2:].reset_index(drop=True)

df2=pd.read_csv('../data/2810/p3-case1.csv')
prt2=df2.iloc[10:,2:].reset_index(drop=True)

df3=pd.read_csv('../data/2810/p4-case1.csv')
prt3=df3.iloc[10:,2:].reset_index(drop=True)


p0=df.iloc[5,2:].values.astype(float)    # Point0
p1=df.iloc[6,2:].values.astype(float)    # Point1
u0=df.iloc[8,2:].values.astype(float)    # Unit Vector @ Point0
u1=df.iloc[9,2:].values.astype(float)    # Unit Vector @ Point1
m0=df.iloc[2,1]                          # Magnitude for Vector1
m1=df.iloc[3,1]                          # Magnitude for Vector2
u0_init=u0*m0
u1_init=u1*m1

utils.plotter(prt.values,p0,p1,u0_init,u1_init,tgt=prt1.values,prt=prt2.values,siml=prt3.values,cne=6,intr='d',tgt_label='Optimized hermite spline')
```
