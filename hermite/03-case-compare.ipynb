### Applied the BCs from 3 cases and comparing with spline data from CAD software

## Case 01

```python
import utils
import pandas as pd

df=pd.read_csv('../data/case-01.csv')
cad=df.iloc[10:,2:].reset_index(drop=True)

p0=df.iloc[5,2:].values                  # Point0
p1=df.iloc[6,2:].values                  # Point1
u0=df.iloc[8,2:].values                  # Unit Vector @ Point0
u1=df.iloc[9,2:].values                  # Unit Vector @ Point1
m0=df.iloc[2,1]                          # Magnitude for Vector1
m1=df.iloc[3,1]                          # Magnitude for Vector2
u0=u0*m0
u1=u1*m1
spl=utils.hermite(p0, p1, u0, u1,num_points=51)

print(f'Length of the curve: {utils.length(spl.astype(float)):.5f}')

utils.plotter(spl,p0,p1,u0,u1,cad=cad.values,padd=20)
```
