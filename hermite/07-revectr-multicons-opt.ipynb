# Position 1

## P1 Case 01

```python
import utils
from scipy.optimize import minimize
import pandas as pd
import numpy as np

df=pd.read_csv('../data/2810/p1-case1.csv')
prt=df.iloc[10:,2:].reset_index(drop=True)

p0=df.iloc[5,2:].values.astype(float)    # Point0
p1=df.iloc[6,2:].values.astype(float)    # Point1
v0=-1*(p0-prt.iloc[3].values).astype(float)
v1=-1*(prt.iloc[96].values-p1).astype(float)
u0=v0/np.linalg.norm(v0)                 # Unit Vector @ Point0
u1=v1/np.linalg.norm(v1)                 # Unit Vector @ Point1
m0=df.iloc[2,1]                          # Magnitude for Vector1
m1=df.iloc[3,1]                          # Magnitude for Vector2
u0_init=u0*m0
u1_init=u1*m1
spl=utils.hermite(p0,p1,u0_init,u1_init,num_points=51)

tgt=utils.length(prt.values)             # Target spline length

def obj_fn(mag,prt_pts,p0,p1,u0,u1,n_points=100):
    m0=mag[0]
    m1=mag[1]    
    u0_scld=m0*u0
    u1_scld=m1*u1
    spl=utils.hermite(p0,p1,u0_scld,u1_scld,num_points=n_points)
    dist_err=abs(np.sum((spl-prt_pts)**2))
    return dist_err

def len_up(mag,prt_pts,p0,p1,u0,u1,n_points=100):
    return 2-(utils.length(utils.hermite(p0,p1,mag[0]*u0,mag[1]*u1,num_points=n_points))-utils.length(prt_pts))

def len_lo(mag,prt_pts,p0,p1,u0,u1,n_points=100):
    return 2+(utils.length(utils.hermite(p0,p1,mag[0]*u0,mag[1]*u1,num_points=n_points))-utils.length(prt_pts))

def len_eq(mag,prt_pts,p0,p1,u0,u1,n_points=100):
    return (utils.length(utils.hermite(p0,p1,mag[0]*u0,mag[1]*u1,num_points=n_points))-utils.length(prt_pts))

constraints=[
    {'type':'eq','fun':len_eq,'args':(prt.values,p0,p1,u0,u1)},
    # {'type':'ineq','fun':len_up,'args':(prt.values,p0,p1,u0,u1)},
    # {'type':'ineq','fun':len_lo,'args':(prt.values,p0,p1,u0,u1)}
]

sol=minimize(obj_fn,[1,1],args=(prt.values,p0,p1,u0,u1),method='SLSQP',constraints=constraints)
u0_final=sol.x[0]*u0
u1_final=sol.x[1]*u1
spl=utils.hermite(p0,p1,u0_init,u1_init,num_points=100)
spl_tgt=utils.hermite(p0,p1,u0_final,u1_final,num_points=100)

pd.DataFrame(spl_tgt, columns=['X','Y','Z']).to_csv('../data/2810/rev-opt/p1-case1-opt.csv',index=False)

print(f'Length of the Prototype Spline: {utils.length(prt.values):.5f}')
print(f'Given target spline length: {tgt:.5f}')
print(f'Length of the Hermite Spline: {utils.length(spl.astype(float)):.5f}')
print(f'Achieved Length of the Hermite Spline: {utils.length(spl_tgt.astype(float)):.5f}')
print(f'\nMaximum Distance Deviation: {max(np.linalg.norm(spl_tgt-spl,axis=1)):.4f}')
print(f'Total Distance Deviation: {sum(np.linalg.norm(spl_tgt-spl,axis=1)):.4f}')
print(f'\nComputed Magnitude (M0, M1): {sol.x[0]:.4f}, {sol.x[1]:.4f}')

utils.plotter(spl,p0,p1,u0_init,u1_init,tgt=spl_tgt,prt=prt.values,cne=6,intr='d',tgt_label='Optimized hermite spline')
```
